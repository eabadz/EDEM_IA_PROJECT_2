## Loan Dataset Features

| Variable                         | Description                                                                                 | Data Type     |
|----------------------------------|---------------------------------------------------------------------------------------------|---------------|
| `person_age`                     | Person's age                                                                                | Float         |
| `person_gender`                  | Person's gender                                                                             | Categorical   |
| `person_education`              | Highest education level                                                                     | Categorical   |
| `person_income`                  | Annual income                                                                              | Float         |
| `person_emp_exp`                 | Years of work experience                                                                   | Integer       |
| `person_home_ownership`          | Home ownership status (e.g., rent, own, mortgage)                                          | Categorical   |
| `loan_amnt`                      | Requested loan amount                                                                      | Float         |
| `loan_intent`                    | Loan purpose                                                                               | Categorical   |
| `loan_int_rate`                  | Loan interest rate                                                                         | Float         |
| `loan_percent_income`            | Loan amount as a percentage of annual income                                               | Float         |
| `cb_person_cred_hist_length`     | Length of credit history in years                                                          | Float         |
| `credit_score`                   | Person's credit score (The higher the score, the lower the assumed risk of default)        | Integer       |
| `previous_loan_defaults_on_file` | Indicator of previous loan defaults                                                        | Categorical   |
| `loan_status` (label)           | Loan approval status: 1 = approved; 0 = rejected                                            | Integer       |

```python
# Crear una nueva columna para los rangos de edad
data['age_range'] = pd.cut(data['person_age'], bins=range(0, int(data['person_age'].max()) + 5, 5))

# Calcular estadísticas por rangos de edad
# Crear una nueva columna para los rangos de años trabajados
data['emp_exp_range'] = pd.cut(data['person_emp_exp'], bins=range(0, int(data['person_emp_exp'].max()) + 5, 5))

# Calcular estadísticas por rangos de años trabajados
emp_exp_range_stats = data.groupby('emp_exp_range').agg({
    'person_income': ['mean', 'median', 'min', 'max'],
    'loan_amnt': ['mean', 'median', 'min', 'max'],
    'loan_int_rate': ['mean', 'median', 'min', 'max'],
    'credit_score': ['mean', 'median', 'min', 'max']
}).reset_index()

print(emp_exp_range_stats)

# Graficar las estadísticas por rangos de años trabajados
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

sns.barplot(x='emp_exp_range', y=('person_income', 'mean'), data=emp_exp_range_stats, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Mean Person Income by Employment Experience Range')
axes[0, 0].set_xlabel('Employment Experience Range')
axes[0, 0].set_ylabel('Mean Person Income')

sns.barplot(x='emp_exp_range', y=('loan_amnt', 'mean'), data=emp_exp_range_stats, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('Mean Loan Amount by Employment Experience Range')
axes[0, 1].set_xlabel('Employment Experience Range')
axes[0, 1].set_ylabel('Mean Loan Amount')

sns.barplot(x='emp_exp_range', y=('loan_int_rate', 'mean'), data=emp_exp_range_stats, ax=axes[1, 0], palette='viridis')
axes[1, 0].set_title('Mean Loan Interest Rate by Employment Experience Range')
axes[1, 0].set_xlabel('Employment Experience Range')
axes[1, 0].set_ylabel('Mean Loan Interest Rate')

sns.barplot(x='emp_exp_range', y=('credit_score', 'mean'), data=emp_exp_range_stats, ax=axes[1, 1], palette='viridis')
axes[1, 1].set_title('Mean Credit Score by Employment Experience Range')
axes[1, 1].set_xlabel('Employment Experience Range')
axes[1, 1].set_ylabel('Mean Credit Score')

plt.tight_layout()
plt.show()
```

**MODELO**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Definimos la columna objetivo
target_column = 'loan_status'  # Changed to string
# Separamos features y target
X = data.drop(columns=[target_column])
y = data[target_column]
```

```python
# One-hot encoding
X = pd.get_dummies(X, drop_first=True)
```

```python
# 1. Dividir en entrenamiento + test
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.15, 
                                                    random_state=42, 
                                                    stratify=y)

# 2. Del entrenamiento, sacar validación
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, 
                                                  test_size=0.15, 
                                                  random_state=42, 
                                                  stratify=y_train)

print("Tamaño entrenamiento:", X_train.shape)
print("Tamaño validación:", X_val.shape)
print("Tamaño test:", X_test.shape)
```

```python
scaler = StandardScaler()

# Ajustamos el scaler SOLO con datos de entrenamiento
scaler.fit(X_train)

# Transformamos
X_train_scaled = scaler.transform(X_train)
X_val_scaled   = scaler.transform(X_val)
X_test_scaled  = scaler.transform(X_test)
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

# 1. Definir el modelo
model = Sequential()

# Capa de entrada + primera capa oculta
model.add(Dense(units=32, activation='relu', input_shape=(X_train.shape[1],)))
# Añade dropout si ves sobreajuste
model.add(Dropout(0.2))

# Segunda capa oculta (opcional si necesitas más profundidad)
model.add(Dense(units=16, activation='relu'))
model.add(Dropout(0.2))

# Capa de salida para binario -> 1 neurona con activación sigmoid
model.add(Dense(1, activation='sigmoid'))

# 2. Compilar el modelo
model.compile(
    loss='binary_crossentropy', 
    optimizer=Adam(learning_rate=0.001), 
    metrics=['accuracy']
)

# 3. Entrenar el modelo
history = model.fit(
    X_train, y_train, 
    validation_data=(X_val, y_val), 
    epochs=30, 
    batch_size=32
)

# 4. Evaluar el modelo
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_acc}")
```
